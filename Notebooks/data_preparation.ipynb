```python
import pandas as pd

melb_raw = pd.read_csv('melb_data.csv')
melb_raw
```

 ### Eliminate rows containing NaN values ###

```python
melb_df = melb_raw.dropna(axis = 0)
melb_df
```

### Save the dataframe ### 

```python
melb_df.to_csv('melb_df.csv', index=False)
```

```python
df = pd.read_csv('melb_df.csv')
df
```

```python
import matplotlib.pyplot as plt
value_counts = df['Price'].value_counts()

plt.hist(sorted(df['Price']),bins=10,color='blue', edgecolor='black',weights=sorted(df['Price']))
plt.title('')
plt.xlabel('Frequency')
plt.ylabel('Prices')

bin_edges = plt.gca().patches  # Get the rectangles that make up the bars
for patch in bin_edges:
    height = patch.get_height()  # Height of the bar
    x_position = patch.get_x() + patch.get_width() / 2  # Center of the bar
    if height > 0:  # Only annotate non-zero heights
        plt.text(x_position, height + 0.5, f'{int(height)}', ha='center', fontsize=10)
plt.show()
# print(len([i for i in df['Price'] if i not in ]))

```

```python
import seaborn as sns
import numpy as np

plt.figure(figsize=(8, 6))
bin_edges = np.arange(0, max(df['Price']), 500000)

sns.histplot(data=df, x='Price', bins=bin_edges, color='#F884C2', edgecolor='black', kde=False)

plt.title('Histogram of Prices', fontsize=16)
plt.xlabel('Price (In Millions)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
# Show the plot
plt.show()
```

```python
corr_matrix = df[['Price', 'Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add a title
plt.title('Correlation Heatmap')

# Show the heatmap
plt.show()
```

```python
df_sp = df[(df['Landsize'] < 5000) & (df['Price'] < 1e7)]
sns.scatterplot(data=df,y='Price',x='Landsize')
plt.title('Square footage VS Price')
plt.show()
```

```python
bins = [0, 500, 1000, 2000, 5000, 10000]  # Adjust the ranges as needed
labels = ['0-500', '501-1000', '1001-2000', '2001-5000', '5001-10000']
df['Landsize_Range'] = pd.cut(df['Landsize'], bins=bins, labels=labels)

# Plot using the new range column
# sns.boxplot(data=df, x='Landsize_Range', y='Price', hue='Type')
sns.scatterplot(data=df, x='Landsize_Range', y='Price', hue='Type')
plt.title('Price vs Landsize Ranges')
plt.xlabel('Landsize (sq ft)')
plt.ylabel('Price')
plt.show()
```
