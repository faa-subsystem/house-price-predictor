# Data Preparation #

```python
import pandas as pd

melb_raw = pd.read_csv('melb_data.csv')
melb_raw
```

 ### Eliminate rows containing NaN values ###

```python
melb_df = melb_raw.dropna(axis = 0)
melb_df
```

### Save the dataframe ### 

```python
melb_df.to_csv('melb_df.csv', index=False)
```
